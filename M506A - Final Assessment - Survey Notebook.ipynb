```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
from sklearn.metrics import confusion_matrix
```

```python
data = {"P1":[1,4,5,5,2,2,2,3,1,3], "P2":[3,4,3,3,4,4,4,4,3,5], "P3":[4,5,1,1,5,5,5,5,5,5], "P4":[2,3,4,4,3,3,3,4,2,4], "P5":[1,2,5,5,4,2,2,3,1,2], "P6":[1,2,5,5,1,2,1,2,1,2], "P7":[5,4,1,1,5,5,5,5,5,5], "P8":[3,4,3,3,4,4,3,3,4,4], "P9":[1,3,5,5,3,2,2,2,1,2], "P10":[5,4,1,1,1,5,5,5,5,5]}

questions = ["Q1 Vocabulary & Grammar", "Q2 Pronunciation", "Q3 Nervous Speaking", "Q4 Afraid of Mistakes", "Q5 Start Conversations", "Q6 Communicate Locally", "Q7 Social Connection", "Q8 Social Activities", "Q9 Speaking Confidence", "Q10 Confidence Improved"]

df = pd.DataFrame(data, index=questions)

fig = go.Figure(data=go.Heatmap(z=df.values, x=df.columns, y=df.index, colorscale="Blues", zmin=1, zmax=5, text=df.values, texttemplate="%{text}", colorbar=dict(title="Score", tickvals=[1,2,3,4,5], ticktext=["1","2","3","4","5"])))

fig.update_layout(title="Survey Questionnaire Responses (Participants vs Questions)", xaxis_title="Participants", yaxis_title="Survey Questions", yaxis=dict(autorange="reversed"))

fig.show()
```

```python
#Survey Data
data = {"P1":[1,4,5,5,2,2,2,3,1,3], "P2":[3,4,3,3,4,4,4,4,3,5], "P3":[4,5,1,1,5,5,5,5,5,5], "P4":[2,3,4,4,3,3,3,4,2,4], "P5":[1,2,5,5,4,2,2,3,1,2], "P6":[1,2,5,5,1,2,1,2,1,2], "P7":[5,4,1,1,5,5,5,5,5,5], "P8":[3,4,3,3,4,4,3,3,4,4], "P9":[1,3,5,5,3,2,2,2,1,2], "P10":[5,4,1,1,1,5,5,5,5,5]}

questions = ["Q1 Vocabulary", "Q2 Pronunciation", "Q3 Nervous Speaking", "Q4 Afraid of Mistakes", "Q5 Start Conversations", "Q6 Communicate Locally", "Q7 Social Connection", "Q8 Social Activities", "Q9 Speaking Confidence", "Q10 Confidence Improved"]

df = pd.DataFrame(data, index=questions)

#Selected two questions
actual = df.loc["Q3 Nervous Speaking"]
predicted = df.loc["Q9 Speaking Confidence"]

#Confusion matrix
cm = confusion_matrix(actual, predicted, labels=[1,2,3,4,5])

#Plotted heatmap
fig = px.imshow(cm, text_auto=True, color_continuous_scale="Blues", labels=dict(x="Speaking Confidence", y="Nervous Speaking", color="Count"), x=[1,2,3,4,5], y=[1,2,3,4,5], title="Confusion Matrix: Nervous Speaking vs Speaking Confidence")

#Removed decimals from color bar
fig.update_coloraxes(colorbar=dict(tickvals=[0,1,2,3,4,5], ticktext=["0","1","2","3","4","5"]))

fig.show()
```
